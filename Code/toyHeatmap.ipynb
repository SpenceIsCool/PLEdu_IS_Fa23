```python
class Data:

    def join(self, that, name):
        Data(name,
                self.create + that.create,
                self.evaluate + that.evaluate,
                self.analyze + that.analyze,
                self.apply + that.apply,
                self.understand + that.understand,
                self.remember + that.remember)

    def __init__(self, name, create, evaluate, analyze, apply, understand, remember):
        self.name = name
        self.create = create
        self.evaluate = evaluate
        self.analyze = analyze
        self.apply = apply
        self.understand = understand
        self.remember = remember

    def __repr__(self):
        return "Name: {name}\n\t- create: {create}\n\t- evaluate: {evaluate} \n\t- analyze: {analyze} \n\t- apply: {apply} \n\t- understand: {understand} \n\t- remember: {remember}".format( name=self.name, create=self.create, evaluate=self.evaluate, analyze=self.analyze, apply=self.apply, understand=self.understand, remember=self.remember)
    
    def graph(self):
        seaborn.heatmap(data, *, vmin=None, vmax=None, cmap=None, center=None, robust=False, annot=None, fmt='.2g', annot_kws=None, linewidths=0, linecolor='white', cbar=True, cbar_kws=None, cbar_ax=None, square=False, xticklabels='auto', yticklabels='auto', mask=None, ax=None, **kwargs)
```
