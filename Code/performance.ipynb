```python
from enum import Enum
```

```python
class Bloom(Enum):
    REMEMBER=1
    UNDERSTAND=2
    APPLY=3
    ANALYZE=4
    EVALUATE=5
    CREATE=6
    def default():
        return Bloom.REMEMBER
    def inc(self):
        match (self):
            case 6: return Bloom.CREATE
            case 5: return Bloom.CREATE
            case 4: return Bloom.EVALUATE
            case 3: return Bloom.ANALYZE
            case 2: return Bloom.APPLY
            case 1: return Bloom.UNDERSTAND
    def dec(self):
        match (self):
            case 6: return Bloom.EVALUATE
            case 5: return Bloom.ANALYZE
            case 4: return Bloom.APPLY
            case 3: return Bloom.UNDERSTAND
            case _: return Bloom.REMEMBER    
```

```python
import random
```

```python
class StudentP2P:
    bloomMidterm=Bloom.default()
    midtermRaw=0
    bloomFinal=Bloom.default()
    finalRaw=0
    def __init__(self, bloomMidterm, bloomFinal):
        self.bloomMidterm = bloomMidterm
        self.setRawMidterm()
        self.bloomFinal = bloomFinal
        self.setRawFinal()
    def findRawExam(self, bloom, avg):
        res = 0
        r1 = random.random()
        r2 = random.random()
        if (r1 < .1):
            match (bloom):
                case Bloom.CREATE: res = random.randrange(90,100)
                case Bloom.EVALUATE: res = random.randrange(avg-10,100)
                case Bloom.ANALYZE: res = random.randrange(avg-20,90)
                case Bloom.APPLY: res = random.randrange(avg-30,80)
                case Bloom.UNDERSTAND: res = random.randrange(avg-40,80)
                case Bloom.REMEMBER: res = random.randrange(avg-50,80)
        elif (r1 < .9):
            match (bloom):
                case Bloom.CREATE: res = random.randrange(avg-5,100)
                case Bloom.EVALUATE: res = random.randrange(avg-10,90)
                case Bloom.ANALYZE: res = random.randrange(avg-20,80)
                case Bloom.APPLY: res = random.randrange(avg-30,70)
                case Bloom.UNDERSTAND: res = random.randrange(avg-40,60)
                case Bloom.REMEMBER: res = random.randrange(avg-50,50)
        else:
            match (bloom):
                case Bloom.CREATE: res = random.randrange(0,avg+5)
                case Bloom.EVALUATE: res = random.randrange(avg-10,90)
                case Bloom.ANALYZE: res = random.randrange(avg-20,80)
                case Bloom.APPLY: res = random.randrange(avg-30,70)
                case Bloom.UNDERSTAND: res = random.randrange(avg-40,60)
                case Bloom.REMEMBER: res = random.randrange(avg-50,50)
        if (res > avg + 10): res = Bloom.CREATE
        elif (res > avg): res = Bloom.EVALUATE
        elif (res > avg - 5): res = Bloom.ANALYZE
        elif (res > avg - 10): res = Bloom.APPLY
        elif (res > avg - 20): res= Bloom.UNDERSTAND
        else: res = Bloom.REMEMBER
        return res
    def setRawMidterm(self):
        self.midtermRaw = self.findRawExam(self.bloomMidterm, 70)
    def setRawFinal(self):
        self.finalRaw = self.findRawExam(self.bloomMidterm, 50)
    def default():
        return StudentP2P(Bloom.default(), Bloom.default())
    def makeRandP2P():
        # randomly from an expected distrobusion set the base...
        # than randomly from another expected distrobution set the final...
        r1 = random.random()        
        midtermBloom = Bloom.default()
        if (r1 < .05): midtermBloom = Bloom.CREATE
        elif (r1 < .2): midtermBloom = Bloom.EVALUATE
        elif (r1 < .8): midtermBloom = Bloom.ANALYZE
        elif (r1 < .9): midtermBloom = Bloom.APPLY
        elif (r1 < .97): midtermBloom = Bloom.UNDERSTAND
        else: midtermBloom = Bloom.REMEMBER

        r1 = random.random()        
        finalBloom = Bloom.default()
        if (r1 < .1): finalBloom = Bloom.CREATE
        elif (r1 < .7): finalBloom = Bloom.EVALUATE
        elif (r1 < .9): finalBloom = Bloom.ANALYZE
        elif (r1 < .96): finalBloom = Bloom.APPLY
        elif (r1 < .99): finalBloom = Bloom.UNDERSTAND
        else: finalBloom = Bloom.REMEMBER
        
        return StudentP2P(midtermBloom, finalBloom)
    
    def __str__(self):
        return "Midterm (%s, %s) Final (%s %s)" % (self.bloomMidterm, self.midtermRaw, self.bloomFinal, self.finalRaw)
```

```python
class StudentTA:
    bloomMidterm=Bloom.default()
    midtermRaw=0
    bloomFinal=Bloom.default()
    finalRaw=0
    def __init__(self, bloomMidterm, bloomFinal):
        self.bloomMidterm = bloomMidterm
        self.setRawMidterm()
        self.bloomFinal = bloomFinal
        self.setRawFinal()
    def findRawExam(self, bloom, avg):
        res = 0
        r1 = random.random()
        r2 = random.random()
        if (r1 < .1):
            match (bloom):
                case Bloom.CREATE: res = random.randrange(90,100)
                case Bloom.EVALUATE: res = random.randrange(avg-10,100)
                case Bloom.ANALYZE: res = random.randrange(avg-20,90)
                case Bloom.APPLY: res = random.randrange(avg-30,80)
                case Bloom.UNDERSTAND: res = random.randrange(avg-40,80)
                case Bloom.REMEMBER: res = random.randrange(avg-50,80)
        elif (r1 < .5):
            match (bloom):
                case Bloom.CREATE: res = random.randrange(avg-5,100)
                case Bloom.EVALUATE: res = random.randrange(avg-10,90)
                case Bloom.ANALYZE: res = random.randrange(avg-20,80)
                case Bloom.APPLY: res = random.randrange(avg-30,70)
                case Bloom.UNDERSTAND: res = random.randrange(avg-40,60)
                case Bloom.REMEMBER: res = random.randrange(avg-50,50)
        else:
            match (bloom):
                case Bloom.CREATE: res = random.randrange(0,avg+5)
                case Bloom.EVALUATE: res = random.randrange(avg-10,90)
                case Bloom.ANALYZE: res = random.randrange(avg-20,80)
                case Bloom.APPLY: res = random.randrange(avg-30,70)
                case Bloom.UNDERSTAND: res = random.randrange(avg-40,60)
                case Bloom.REMEMBER: res = random.randrange(avg-50,50)
        return res/100
    def setRawMidterm(self):
        self.midtermRaw = self.findRawExam(self.bloomMidterm, 70)
        # spwi random average centerd at 70 (div by 100)
    def setRawFinal(self):
        self.finalRaw = self.findRawExam(self.bloomFinal, 50)
    def default():
        return StudentP2P(Bloom.default(), Bloom.default())
    def makeRandTA():
        # randomly from an expected distrobusion set the base...
        # than randomly from another expected distrobution set the final...
        r1 = random.random()        
        midtermBloom = Bloom.default()
        if (r1 < .01): midtermBloom = Bloom.CREATE
        elif (r1 < .02): midtermBloom = Bloom.EVALUATE
        elif (r1 < .11): midtermBloom = Bloom.ANALYZE
        elif (r1 < .6): midtermBloom = Bloom.APPLY
        elif (r1 < .9): midtermBloom = Bloom.UNDERSTAND
        else: midtermBloom = Bloom.REMEMBER

        r1 = random.random()        
        finalBloom = Bloom.default()
        if (r1 < .01): finalBloom = Bloom.CREATE
        elif (r1 < .05): finalBloom = Bloom.EVALUATE
        elif (r1 < .15): finalBloom = Bloom.ANALYZE
        elif (r1 < .75): finalBloom = Bloom.APPLY
        elif (r1 < .95): finalBloom = Bloom.UNDERSTAND
        else: finalBloom = Bloom.REMEMBER
        
        return StudentP2P(midtermBloom, finalBloom)
    
    def __str__(self):
        return "Midterm (%s, %s) Final (%s %s)" % (self.bloomMidterm, self.midtermRaw, self.bloomFinal, self.finalRaw)
```
